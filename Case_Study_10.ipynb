# Probablity

```python
# Importing the necessary libraries
import  numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the data to Python environment
data = pd.read_csv('mushrooms.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
# Checking for Null values in the data set
data.isnull().sum()
```

There are no null values present in this data set.

```python
data.describe(include = 'object')
```

```python
# since'veil-type'only have one unique value, we can remove this column.
data = data.drop(["veil-type"],axis=1)
```

```python
# Spliting the data
X = data.drop(['class'],axis =1)
y = data['class']
X = pd.get_dummies(X)
X.head()
```

```python
#Label encoding 
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder() 
y = label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.3)
```

# 1. Logistic Regression Classification

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver="lbfgs")
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

lr_as = accuracy_score(y_test,y_pred)
lr_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

# 2.K-Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
    
plt.plot(k_range, scores,marker="o")
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of K-Nearest-Neighbors')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

knn_as = accuracy_score(y_test,y_pred)
knn_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

# 3. SVM Classifier

```python
from sklearn.svm import SVC

svm = SVC(random_state=42, gamma="auto")
svm.fit(X_train,y_train)

y_pred = svm.predict(X_test)

svm_as = accuracy_score(y_test,y_pred)
svm_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

# 4. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train,y_train)

y_pred = rf.predict(X_test)

rf_as = accuracy_score(y_test,y_pred)
rf_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

# 5. Naive Bayes Classifier

## A) Gaussian Naive Bayes Classifier 

```python
from sklearn.naive_bayes import GaussianNB

gauss_nb = GaussianNB()
gauss_nb.fit(X_train,y_train)

y_pred = gauss_nb.predict(X_test)

nb1_as = accuracy_score(y_test,y_pred)
nb1_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

## B) Bernoulli Naive Bayes Classifier

```python
from sklearn.naive_bayes import BernoulliNB

ber_nb = BernoulliNB()
ber_nb.fit(X_train,y_train)

y_pred = ber_nb.predict(X_test)

nb2_as = accuracy_score(y_test,y_pred)
nb2_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

## C) Categorical Naive Bayes 

```python
from sklearn.naive_bayes import CategoricalNB

cat_nb = CategoricalNB()
cat_nb.fit(X_train,y_train)

y_pred = cat_nb.predict(X_test)

nb3_as = accuracy_score(y_test,y_pred)
nb3_f1 = f1_score(y_test,y_pred)

confusion_matrix(y_test,y_pred)
```

```python
# Comparing the Accuracy and F-1 Score
```
